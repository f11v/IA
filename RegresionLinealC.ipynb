```python
# Importar Librerias 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importar el Dataset desde el enlace proporcionado
url = 'https://raw.githubusercontent.com/alavi-sam/salary-prediction/main/Salary%20Data.csv'
dataset = pd.read_csv(url)

# Eliminar filas con valores NaN
dataset = dataset.dropna()

# Seleccionar variables: 'YearsExperience' como independiente (X) y 'Salary' como dependiente (y)
X = dataset[['Years of Experience']].values  # Años de experiencia
y = dataset['Salary'].values  # Salario

# División de los datos en conjunto de entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

# Crear el modelo de regresión lineal
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train, y_train)

# Predicción de los resultados
y_pred = regression.predict(X_test)

# Visualización de los resultados para el conjunto de entrenamiento
plt.scatter(X_train, y_train, color="red")
plt.plot(X_train, regression.predict(X_train), color="blue")
plt.title("Salario vs. Experiencia (Entrenamiento)")
plt.xlabel("Años de Experiencia")
plt.ylabel("Salario")
plt.show()

# Visualización de los resultados para el conjunto de prueba
plt.scatter(X_test, y_test, color="red")
plt.plot(X_test, regression.predict(X_test), color="blue")
plt.title("Salario vs. Experiencia (Prueba)")
plt.xlabel("Años de Experiencia")
plt.ylabel("Salario")
plt.show()

# Imprimir coeficientes
print(f"Coeficiente de pendiente (m): {regression.coef_[0]}")
print(f"Intersección (b): {regression.intercept_}")
```
